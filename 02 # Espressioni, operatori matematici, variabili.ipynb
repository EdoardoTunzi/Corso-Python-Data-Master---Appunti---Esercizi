### Operatori aritmetici
Ricordiamo
che usiamo Jupyter, ovvero che stiamo operando in un ambiente interattivo,
pensato per eseguire operazioni di qualsiasi
tipo, come ad esempio anche le operazioni numeriche,
chiamate *espressioni* in Python.

Scrivete ad esempio ```2 + 2``` e premete shift+invio.
Esce il risultato ```4``` di un'espressione che ha coinvolto
**valori** (i numeri) e gli **operatori** (il +).

In caso di espressioni numeriche più complesse, l'ordine
di esecuzione segue lo standard matematico moltiplicazioni
e divisioni per prime, con addizioni e sottrazioni a seguire.
Scrivete ```1 + 2 * 3``` e premete SHIFT+INVIO. Verrà svolta prima la
moltiplicazione e poi l'addizione anche se sono scritte in
ordine inverso. A questo punto scrivete ```(1 + 2) * 3``` per
far vedere come possiamo impostare l'ordine tramite l'utilizzo
delle parentesi.

````python
8 / 2
````

in questi esempi abbiamo utilizzato gli **interi**, uno dei tipi
di dato disponibili in Python. I numeri non interi o in virgola
mobile sono chiamati **float**. Ad esempio ```3 / 2``` darà in
output un valore di tipo float, ovvero ```1.5``` (specificare che
si usa il punto come separatore decimale).

*elencare gli operatori matematici di Python con brevi esempi*
```1 + 1``` addizione
```2 - 1``` sottrazione
```3 * 2``` mopltiplicazione
```3 / 2``` divisione
```2 ** 3``` due alla terza che dà 8
```20 // 3``` divisione intera che dà 3
```20 % 3``` resto di una divisione che da 2

````python
2 * ((3 + 5) - 7) ** 2
````

### Variabili
possiamo associare un valore ad una **variabile**,
pensandola come il modo in cui un computer riesce
a memorizzare dei valori, li mette in una "scatola"
che ha un nome unico nel contesto di un programma.
Con la sintassi
```x = 42``` assegnamo (si chiama assegnazione questo
processo) il valore 42 alla variabile ```x```
che possiamo riutilizzare in seguito.

Usare la funzione ```type``` per farsi restituire
il tipo di una variabile (digitare) ```type(x)``` (premere shift+invio)
esce ```class 'int'``` ad indicare che ```x``` è di tipo ```int```.

scrivere ```x = 42.0``` e rifare ```type(x)``` per verificare che
sia diventato un float. La "scatola" ```x``` può contenere
un solo valore alla volta e quanto gli viene assegnato un nuovo
valore, dimentica il precedente per poter ricordare il nuovo.

dire che Python seleziona la tipologia di dato più consona a
secondo del valore che gli viene passato

scrivere ```testo = "ciao"``` e dire che i testi sono un
formato di dato chiamato ```string``` (stringhe, sequenze
di caratteri alfanumerici)

far presente che in python possiamo usare lettere, numeri e
underscore _(trattino basso, specificare)_ per i nomi delle
variabili ma non possiamo creare un nome di variabile che
inizia con un numero, darebbe errore di sintassi.

suggerire che è importante dare nomi specifici alle variaibili,
relativamente al contesto ed al motivo per cui sono usate.

tornando alle stringhe, se sono state definite con il carattere
```"```, dire che si deve mettere un backslash prima di un
eventuale altro carattere ```"``` all'interno della stringa di testo.
scrivere ```print("e Roberta disse \"Oggi sono proprio contento!\"")```
commentando i doppi doppi apici alla fine della stringa ;)

a questo punto vediamo come l'operatore "+" che abbiamo usato in precedenza per effettuare operazioni aritmetiche possa essere usato con operandi di tipo diverso, ad esempio per una concatenazione di stringhe
utilizzando le variabili:
