## Esercizio 1 — Congettura di Goldbach

La congettura di Goldbach dice che ogni numero pari maggiore di 2 si può scrivere come somma di due numeri primi. Il programma riceve un numero pari `n` e stampa tutte le coppie che soddisfano questa condizione.

L'idea è semplice: proviamo tutti i valori `i` da 2 fino a `n/2`. Per ognuno calcoliamo `j = n - i` e verifichiamo se sia `i` che `j` sono primi. Ci fermiamo a metà perché oltre si ripeterebbero le stesse coppie in ordine inverso.

Per verificare se un numero è primo, usiamo un ciclo `for` interno che prova tutti i possibili divisori da 2 fino al numero stesso: se ne trova uno che divide esattamente, il numero non è primo.

````python
n = 28  # prova a cambiare questo valore con un altro numero pari

for i in range(2, n // 2 + 1):

    # verifichiamo se i e' primo
    i_primo = True
    for d in range(2, i):
        if i % d == 0:      # d divide i senza resto: i non e' primo
            i_primo = False
            break           # trovato un divisore, inutile andare avanti

    # verifichiamo se j = n - i e' primo
    j = n - i
    j_primo = True
    for d in range(2, j):
        if j % d == 0:
            j_primo = False
            break

    # se entrambi sono primi, la coppia e' valida
    if i_primo and j_primo:
        print(f"{i} + {j} = {n}")
````

Con `n = 28` l'output sarà:

```
5 + 23 = 28
11 + 17 = 28
```

Il `break` dentro il ciclo di verifica non cambia il risultato, ma evita di continuare a cercare divisori quando ne abbiamo già trovato uno. Con numeri grandi la differenza di tempo si sente.

## Esercizio 2 — Congettura di Collatz

Partiamo da un numero intero positivo e applichiamo sempre la stessa regola: se è pari lo dividiamo per 2, se è dispari lo moltiplichiamo per 3 e aggiungiamo 1. La congettura dice che, qualunque sia il punto di partenza, prima o poi arriveremo a 1.

Ci serve un ciclo `while` perché non sappiamo in anticipo quanti passi ci vorranno. Ad ogni iterazione aggiorniamo tre cose: la stringa che rappresenta la sequenza, un contatore per la lunghezza e una variabile che tiene il valore più alto raggiunto fin lì.

Usiamo una stringa al posto di una lista per accumulare la sequenza, perché le liste non sono ancora state introdotte. È un caso in cui si vede concretamente perché quella struttura dati tornerà utile nelle prossime lezioni.

````python
n = 6  # prova a cambiare questo valore

sequenza = str(n)   # partiamo con il primo elemento gia' nella stringa
lunghezza = 1       # contiamo anche n di partenza
massimo = n         # per ora il massimo e' n stesso

while n > 1:
    if n % 2 == 0:          # n e' pari
        n = n // 2
    else:                   # n e' dispari
        n = 3 * n + 1

    sequenza = sequenza + ", " + str(n)   # aggiungiamo il nuovo valore alla stringa
    lunghezza = lunghezza + 1

    if n > massimo:         # aggiorniamo il massimo solo se serve
        massimo = n

print(f"Sequenza: {sequenza}")
print(f"Lunghezza: {lunghezza}")
print(f"Valore massimo raggiunto: {massimo}")
````

Con `n = 6` l'output sarà:

```
Sequenza: 6, 3, 10, 5, 16, 8, 4, 2, 1
Lunghezza: 9
Valore massimo raggiunto: 16
```

Vale la pena provare numeri diversi: la sequenza può essere sorprendentemente lunga anche per valori piccoli. Il numero 27, ad esempio, richiede 111 passi prima di arrivare a 1.

## Esercizio 3 — Convertitore di Base Manuale

Il problema è convertire un numero decimale in una base qualsiasi tra 2 e 16, senza usare le funzioni già pronte di Python come `bin()` o `hex()`.

Il metodo si chiama "divisioni successive": si divide il numero per la base di destinazione, si prende il resto, poi si continua a dividere il quoziente finché non diventa zero. I resti, letti dall'ultimo al primo, formano il numero nella nuova base.

Per gestire le basi fino a 16 usiamo una stringa come tabella di corrispondenza: `cifre[10]` è `'A'`, `cifre[15]` è `'F'`. Costruiamo il risultato aggiungendo ogni nuova cifra all'inizio della stringa, in modo che alla fine sia già nell'ordine corretto senza doverla invertire.

````python
numero = 255   # numero decimale da convertire
base   = 16    # base di destinazione, tra 2 e 16

# la stringa funziona come tabella: l'indice e' il valore, il carattere e' la cifra
cifre = "0123456789ABCDEF"

if numero == 0:
    risultato = "0"     # lo zero rimane zero in qualsiasi base
else:
    risultato = ""
    n = numero          # copia di lavoro per non perdere il valore originale

    while n > 0:
        resto = n % base                        # cifra meno significativa
        risultato = cifre[resto] + risultato    # aggiungiamo in testa, non in coda
        n = n // base                           # scendiamo al quoziente

print(f"{numero} in base {base} = \"{risultato}\"")
````

Proviamo qualche caso per verificare che tutto funzioni correttamente:

````python
cifre = "0123456789ABCDEF"

test_cases = [(255, 16), (13, 2), (255, 2), (42, 8)]

for numero, base in test_cases:
    if numero == 0:
        risultato = "0"
    else:
        risultato = ""
        n = numero
        while n > 0:
            resto = n % base
            risultato = cifre[resto] + risultato
            n = n // base
    print(f"{numero} in base {base} = \"{risultato}\"")
````

Output atteso:

```
255 in base 16 = "FF"
13 in base 2 = "1101"
255 in base 2 = "11111111"
42 in base 8 = "52"
```

Si può verificare a mano il caso binario: 13 = 8 + 4 + 1 = 2^3 + 2^2 + 2^0, che corrisponde esattamente a `1101`.
