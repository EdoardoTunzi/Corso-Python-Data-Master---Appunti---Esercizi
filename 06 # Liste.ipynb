## Strutture dati

Le strutture dati sono un modo per aggregare ed organizzare dati di tipo primitivo o composto in maniera che siano più facili il loro accesso e la loro manipolazione. 

Il primo esempio di struttura dati che vediamo è la lista. Essa è un tipo built-in  di Python:

```python
numbers = [-2, 5, 10, 2, 1, 29, 7]
print(type(numbers))

l = len(numbers)

print("La lista ha", l, "elementi")
```

Esercizio classico: trovare la media di una lista. Trovare le occorrenze di una stringa in una lista. Metodo split.

```python
average = 0
sum = 0
for i in numbers:
    print("Sommo l'elemento", i)
    sum += i

average = sum / len(numbers)
print(average)
```

```python
s = "Tizio Hello Sempronio Caio Hello Tizio Caio Caio Sempronio Hello Hello"

l = s.split()
print(l)
cont = 0

for i in l:
    if i == "Hello":
        cont = cont + 1

print("la parola 'Hello' appare", cont, " volte")

```
