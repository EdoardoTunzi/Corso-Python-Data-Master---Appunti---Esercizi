## Cicli for

I cicli sono gruppi di istruzioni ripetute finchè una determinata condizione non sopraggiunge. Abbiamo due grandi gruppi di cicli in Python. I cicli for o i cicli while. I primi si usano per ripetere le stesse istruzioni un numero definito e noto a priori di volte. 

Per ciclare un numero N di volte si usa in Python una funzione che si chiama range(min, max, step). Essa crea una serie di numeri che parte da min e si ferma a max - 1 con un incremento pari a "step".

Partiamo subito con un esempio:

```python
for i in range(0, 10):
    print("la variabile i vale ", i)

print("\n") #stampa una riga vuota
for i in range(0, 10, 2):
    print("la variabile i vale ", i)
```

Esempio con range:

Esempio: stampare tutti i numeri pari nell'intervallo -10, 10, comprendendo anche gli estremi. 

```python
min = -10
max = 10

for num in range(min, max + 1):
    print(num)
```

A questo punto sarebbe bene introdurre il break. Esempio: stampare i primi 5 numeri pari trovati nell'intervallo precedente.

```python
cont = 0

for i in range(min, max + 1):
    if i % 2 == 0:
        print(i, "pari")
        cont = cont + 1
    
    if cont >= 5:
        print("Ho trovato i primi 5 numeri pari dell'intervallo, esco")
        break
```

Come ulteriore esercizio: riprendiamo le stringhe. Possiamo iterare gli elementi di una stringa utilizzando il ciclo for. Esercizio: contare le occorrenze di una certa lettera in una stringa.

```python
s = "Lorem ipsum dolor sit amet mamma"
contatore_m = 0
for letter in s:
    if letter == "m":
        contatore_m = contatore_m + 1
print("La lettera 'm' appare", contatore_m, "volte")
```
