Una funzione in generale è un blocco di codice che prende 0 o più argomenti in input e restituisce un output. Può essere vista quindi come una concretizzazione del concetto di "algoritmo", oppure del concetto di funzione matematica, per cui:

$ y = f(x, y) $

in cui `x` e `y` sono i parametri di input e `y`è il valore di output.

Ci sono due fasi fondamentali nel ciclo di vita di una funzione: 
- la sua dichiarazione
- la sua invocazione

Dichiarare una funzione significa fondamentalmente dare un nome ad un blocco di codice che esegue delle istruzioni. Si può dichiarare una funzione con la parola chiave `def`:

```python
def my_func():
    print("Hello")
    print("Hello i'm a function!")
    
print(my_func)
print(my_func())

print(type(my_func))
```

Questa funzione stampa un messaggio utilizzando la classica funzione `print`. Il momento successivo è l'invocazione della funzione. Una funzione si invoca scrivendo il nome della funzione seguito dalle parentesi tonde (chiamata a funzione). Il tipo di dato di una funzione è "function".

```python
def test_return():
    return 4

a = "HELLO"
print(a)
def sum(a, b):
    print("a -> ", a)
    print("b -> ", b)
    return a + b

a = sum(6, 4)
print(a)
```

Questa funzione non fa altro che stampare un messaggio a schermo. Non è molto utile, vero? Una funzione diventa molto più utile se restituisce un dato in output e se prende dei dati in input. Una funzione che restituisce un dato in output deve avere come ultima istruzione una `return` seguita dal valore o dalla variabile che costituirà l'output della funzione. Gli argomenti in input invece sono dichiarati fra le parentesi tonde che si trovano a fianco al nome della funzione stessa. 

## Argomenti posizionali

Se una funzione ha degli argomenti di input nella sua dichiarazione saremo obbligati ad utilizzarli al momento della sua chiamata (a meno di alcuni casi che vedremo fra poco). Gli argomenti di questo tipo sono detti _posizionali_.

```python
a = sum(1, 2)
```

Nell'esempio precedente: 
- `1 -> a`
- `2 -> b`

## Keyword arguments

È possibile dire esplicitamente quali devono essere i parametri passati alle funzioni durante le loro chiamate. Questi parametri sono detti "Keyword arguments"

```python
b = sum(a=7, b=8)
print(b)


c = sum(b=10, a=8)
```

## Default arguments

È possibile specificare dei valori di default per alcuni o tutti i parametri in fase di definizione di una funzione:

```python
def mod(a, b=2):
    print('a ->', a)
    print('b ->', b)  
    return a % b


c = mod(10, 3)
print(c)

c = mod(10)
print(c)
```

Esempio: scrivere una funzione che prenda in input una lista di numeri interi e restituisca una lista formata solo dai numeri pari

```python
def get_even_numbers(l=[]):
    even = []
    for i in l:
        if mod(i) == 0:
            even.append(i)
    return even
```

```python
numbers = [2, 4, 92, 1, 22, 239, 49]

print(get_even_numbers(numbers))
```
