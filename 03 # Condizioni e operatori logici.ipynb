## Tipo booleano

Il tipo booleano è un tipo di dato built-in di Python che può avere solo valori False e True. Come abbiamo detto nella lezione precedente, possiamo assegnare ad una variabile un valore diretto anche per i dati booleani. 

```python
boolean_var = True
print(type(boolean_var))
print(boolean_var)
```

L'assegnazione diretta non è l'unico modo con cui è possibile assegnare ad una variabile un valore booleano. Esistono degli operatori, in Python, che restituiscono non un valore numerico o alfanumerico, bensì un valore booleano. È il caso degli operatori di confronto, che restituiscono (appunto) un valore booleano. La peculiarità di questi operatori è che usano operandi non booleani e restituiscono un risultato booleano.

Gli operatori di confronto in Python sono:
- `==`: operatore di uguaglianza: ci dice se un operando è uguale all'altro
- `!=`: operatore di diversità: ci dice se un operando è diverso dall'altro
- `>`: operatore che confronta se un operando è strettamente maggiore dell'altro
- `>=` : operatore che confronta se un operando è maggiore o uguale dell'altro
- `<`: operatore che confronta se un operando è strettamente minore dell'altro
- `<=` : operatore che confronta se un operando è minore o uguale dell'altro

```python
num = 10

num_2 = 20

boolean_var = num != num_2
print(boolean_var)
```

Un'altra famiglia di operatori è quella degli operatori logici: essi lavorano con operandi booleani e restituiscono un risultato booleano. Gli operatori logici più comuni in Python sono: 

- `and`: restituisce True se entrambi gli operandi sono True
- `or`: restituisce True se almeno un operando è True
- `not`: è un operatore unario che restituisce il reciproco del valore dato in ingresso

```python
negated = not boolean_var
print(negated)

and_result = negated and boolean_var
print(and_result)


or_result = negated or (num != num_2)
print(or_result)
```

È possibile usare anche dei valori non booleani in contesto booleano. Quando una variabile è usata in questo modo si parla di valori *truthy* o *falsy*. Ad esempio, sono *falsy* i valori: 
- stringa vuota
- numero 0
- None (valore speciale)
- tuple, liste, dict vuoti

## if/else e controllo di flusso

Normalmente l'esecuzione di un programma è sequenziale. Ci sono tuttavia dei casi in cui c'è bisogno di saltare l'esecuzione di alcune istruzioni o in cui si deve scegliere un gruppo di istruzioni da eseguire al posto di un altro gruppo. In casi come questi si usa il costrutto if/else:


Sintassi: 

`if <cond>:
    <statement1>
    <statement2>
    `

Esempio: Valorizzare una variabile numerica e stampare un messaggio se è maggiore di 10. 

```python
# n = int(input("Inserisci un numero\t"))
n = 9

if n > 10:
    print("Il numero è maggiore di 10")
    print("Hello")

print("Questa stampa verrà effettuata sempre")
```

### if con elif ed else

```python
if n < 10:
    print("Il numero è minore di 10")
elif n > 10:
    print("Il numero è maggiore di 10")
else:
    print("Il numero è uguale a 10")
```

Come esercizio "live" potremmo verificare se un numero rappresenta un anno bisestile. Per vedere se un anno è bisestile dobbiamo dividerlo per 400; se è divisibile allora è sicuramente un anno bisestile; altrimenti saranno bisestili solo quegli anni che si possono dividere per 4 ma non per 100.


```python
anno = 2000

if anno % 400 == 0:
    print("L'anno ", anno , " è bisestile perché divisibile per 400")
elif anno % 4 == 0 and anno % 100 != 0:
    print("L'anno ", anno ," è bisestile perché divisibile per 4 ma non per 100")
else:
    print("L'anno ", anno , " non è bisestile.")
```
