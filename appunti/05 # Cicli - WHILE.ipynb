Un while è diverso da un for nel senso che in un while non sappiamo a priori quale sarà il numero massimo di iterazioni eseguite. 

Sintassi: while + un'operazione che restituisce un valore booleano. Se il valore booleano è True, le istruzioni vengono eseguite, altrimenti non vengono eseguite. 


```python
n = 5

while n < 10:
    print(n)
    n = n + 1
    
print("La variabile n vale ", n)
```

L'incremento in questo caso è fatto esplicitamente. 

Esempio: usare un ciclo while che stampi la serie di fibonacci fino a 10

```python
x = 0
y = 1

s = 0

while s < 10:
    print(s)
    x = y
    y = s
    s = x + y
```

Altro esercizio: tabellina del 7

```python
tab = 7
molt = 1

while molt <= 10:
    print(tab, " * ", molt, "=", tab*molt)
    molt += 1
```

Bonus: stesso esercizio ma con ciclo for:

```python
for i in range(1, 11):
    print(tab, " * ", i, "=", tab*i)
```
