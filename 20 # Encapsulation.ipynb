```python
class Person:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name
        
    def __str__(self):
        return f"{self.first_name} {self.last_name}"
    

class BankAccountOwner(Person):
    def __init__(self, first_name, last_name, bank_account_id):
        super().__init__(first_name, last_name)
        self.__bank_account_id = bank_account_id
        
    def __str__(self):
        return f"{super().__str__()} con Conto bancario {self.__bank_account_id}"
        
p1 = Person("Giuseppe", "Mastrandrea")
p2 = BankAccountOwner("Giuseppe", "Mastrandrea", "asdfghjk")

print(p1)

p1.first_name = "Francesco"

print(p2)

p2.bank_account_id = "ASDAS"

print(dir(p2))

```
