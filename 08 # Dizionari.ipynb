Un dizionario in Python è un insieme non ordinato di coppie chiave/valore. Le coppie sono separate fra loro da una virgola e sono racchiuse all'interno di due parentesi graffe. Fare esempio dizionario:

```python
car = {
    "marca": "Ford",
    "modello": "Focus"    
}
```

Illustrare la differenza fra i 2 metodi principali di accesso alle proprietà di un dizionario: 

```python
print(car["marca"])
```

```python
print(car.get("marca"))
```

```python
print(car["motore"])
```

```python
print(car.get("Motore", None))
```

Cancellazione e aggiunta di coppie:

```python
del car["marca"]
```

```python
print(car)
```

```python
car["marca"] = "Fiat"
```

```python
car
```

```python
car["motore"] = {
    "alimentazione": "benzina",
    "cilindrata": 1.6,
    "cavalli": 125
}
```

```python
car
```

```python
print(car.get("motore").get("alimentazione"))
```

Creare un dizionario che rappresenti un film e stampare le informazioni relative agli attori, memorizzati come array

```python
movie = {
    "titolo": "C'era una volta il West",
    "regista": "Sergio Leone",
    "attori": ["Claudia Cardinale", "Henry Fonda", "Charles Bronson"],
    "generi": ["Western", "Drammatico"]
}

print("Un film ha queste caratteristiche: \n")

for k in movie.keys():
    print(k)
```

```python
print(movie.get("titolo"), ' - Regia di ', movie.get("regista") )
print("Attori:")
for a in movie.get("attori"):
    print("\t", a)
    
print("Generi:")
for g in movie.get("generi"):
    print("\t", g)
```

Esercizio con lista di film:

```python
movies = [
    {
        "titolo": "C'era una volta il West",
        "regista": "Sergio Leone",
        "attori": ["Claudia Cardinale", "Henry Fonda", "Charles Bronson"],
        "generi": ["Western", "Drammatico"]
    },
    {
        "titolo": "Forrest Gump",
        "regista": "Robert Zemeckis",
        "attori": ["Tom Hanks", "Gary Sinise", "Robin Wright"],
        "generi": ["Drammatico"]
    },
]

for movie in movies:
    print(movie.get("titolo"), ' - Regia di ', movie.get("regista") )
    print("Attori:")
    for a in movie.get("attori"):
        print("\t", a)

    print("Generi:")
    for g in movie.get("generi"):
        print("\t", g)
```
